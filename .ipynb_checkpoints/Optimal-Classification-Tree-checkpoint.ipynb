```python
def calculate_combined_error(dataFrame, targetClass):
    return 1 - (len(dataFrame.query("Class == @targetClass").index)/len(dataFrame.index))
    
# Calculates the error value of how well the tree fits the data
# Input: Tree, Data
# Recursive function
def error_value(tree, dataFrame):
    error = 0
    # misclassification error
    if not tree.is_leaf:
        
        # split data by decision
        feature, operator, value = tree.name.split()
        leftData,rightData  = split_data(feature, operator, value, dataFrame)
        
        # check if node is on left
        if((not leftData.empty) and not tree.children[0].right):
            error += error_value(tree.children[0],leftData)
        if((not rightData.empty) and len(tree.children)>1):
            error += error_value(tree.children[1],rightData)
        elif((not rightData.empty) and tree.children[0].right):
            error += error_value(tree.children[0],rightData)
    else:

        '''
        minimising with respect to tau(.):
        u * sum of (outcome of point if leaf has the same prescription + sum of rest of the means of the other prescriptions)
        + (1-u) * sum of (outcome of point - mean of outcome of same treatment of that leaf)**2

        '''
        
        error = tree.error
    return error



# Calculates the penalizing value for the complexity of the tree
# Input: Tree
# Recursive function
def complexity(tree):
    length = 0
    if not tree.is_leaf:
        length += 1
        for child in tree.children:
            length += complexity(child)
    return length

'''
def complexity(tree):
    length = 0
    if not tree.is_leaf:
        for child in tree.children:
            length = max(complexity(child),length)
    return length + 1

'''


# calculates the loss function for the tree and the respective data
def loss_function(tree, dataFrame, alpha = 0.5):
    return error_value(tree, dataFrame) + alpha * complexity(tree)

#counterfactual(simulatedData,1)
```

```python
# Create a leaf
# Uses most populous class in available data frame
def create_leaf(dataFrame,parent=None,right=None):
    
    # Fix a bug with empty dataframes !!!!!!!!!!!!!!!!
    if not dataFrame.empty:
        # Get the most populous class
        Class = dataFrame['Class'].value_counts().first_valid_index()

        # Get accuracy of class
        accuracy = (dataFrame['Class'].value_counts().iloc[0]/dataFrame['Class'].value_counts().sum()).round(3)
        error = calculate_combined_error(dataFrame, Class)
        # Create Node using previos node's parent
        leaf = Node("class = {Class}, accuracy = {accuracy}, error = {error}".format(Class=Class,accuracy=accuracy, error = error),parent=parent, Class = Class, accuracy = accuracy, error = error, right = right, checked = False)

        return leaf
    else:
        return None
    
# recursive function for creating trees
def create_tree(dataFrame,features,parent = None, length = None, right = False,operators=["<","=="]):
    if length == None:
        length = len(features)
    # create node with splits
    # If there are still features available, create a decision node
    if(length > 0):
        # Remove class from features to pick
        availableFeatures = features[:]
        
        # split decision
        ############################################
        # Pick random feature and operator
        feature, operator = random.choice(availableFeatures), random.choice(operators)
        # Choose random value in available feature
        value = random.choice(dataFrame[feature]).round(3)
        ############################################
        # Create decision node
        
        tree = Node('{feature} {operator} {value}'.format(feature = feature,operator = operator,value = value),parent = parent, right = right, checked = False)
        
        # Split dataset on decision
        
        leftDataFrame, rightDataFrame = split_data(feature, operator, value, dataFrame)
        # create child nodes of decision
        if(not leftDataFrame.empty):
            create_tree(leftDataFrame,[n for n in features if n != feature],tree,length=length-1,right=False)
        if(not rightDataFrame.empty):
            create_tree(rightDataFrame,[n for n in features if n != feature],tree,length=length-1,right=True)
        #tree.children = (create_tree(leftDataFrame,[n for n in features if n != feature],tree,right=False),create_tree(rightDataFrame,[n for n in features if n != feature],tree,right=True))
    # Create a class node
    else:
        # create leaf
        tree = create_leaf(dataFrame, parent, right)
    return tree

# prints the tree in ascii
def print_tree(tree):
    for pre, fill, node in RenderTree(tree):
        print("%s%s, %s" % (pre, node.name,"Left" if not node.right else "Right"))
        
# create test tree
finalTree = create_tree(test_data, get_features(test_data), length = 3)
print_tree(finalTree)
```

```python




#######



# finds deepest node on the left
# for testing purposes
def deepest_node(tree):
    
    if tree.children:
        return deepest_node(tree.children[0])
    else:
        return tree

    
    
# trims the dataset to appropriate data for the node
def data_values(node, dataFrame,right = None):
    # trim data starting from parent value
    
    if node.parent:
        dataFrame = data_values(node.parent,dataFrame,node.right)
    # if data is being split at node and the node has a split
    # Split data according to next node direction
    if right != None and node.children:
        feature, operator, value = node.name.split()
        return split_data(feature, operator, value, dataFrame)[int(right)]
    # If leaf or not splitting data at node
    # Return data at that node
    else:
        return dataFrame

    
    
def gini_single(dataFrame):
    # Lower is better
    length = len(dataFrame)
    gini = 1
    for Class in dataFrame["Class"].unique():
        gini -= float(((len(dataFrame[dataFrame['Class'] == Class].index))/length)**2)
    return gini


    
def optimal_data_split(dataFrame,operators=["<","=="]):
    lowestError = [np.inf,"0","==","0"]#pd.DataFrame(columns=["Gini Value","Feature","Operator","Value"])
    Class = dataFrame['Class'].value_counts().first_valid_index()
    # loops through all data points
    for feature in get_features(dataFrame):
        # sort values in feature in ascending order
        values = dataFrame[feature].drop_duplicates().sort_values().reset_index(drop=True)
        # run test on all 
        for i in range(0, len(values)-1):
            point = (1/2 * (values.iloc[i] + values.iloc[i+1])).round(3)
            newError = 0
            for operator in operators:
                for splitData in split_data(feature,"<",point,dataFrame):
                    if not splitData.empty:
                        newError += gini_single(splitData)
                if newError < lowestError[0]:
                    lowestError = [newError,feature,"<",point]
                    print(lowestError)
                #giniValues.append([newGiniValue,feature,operator,point])
                
    return lowestError[1:4]
  
    
    
#######

def place_left_node(node):
    if (not node.right and node.parent):
        if(len(node.parent.children)>1):
            node.parent.children = [node.parent.children[1],node.parent.children[0]]

# Create new split at given node
def optimal_split(node, dataFrame):
    # create two classification children
    
    # extract set of features
    features = get_features(dataFrame)
    dataFrame = data_values(node,dataFrame)
    if(len(dataFrame.index)) > 1:
        # create split
        feature, operator, value = optimal_data_split(dataFrame)
        parent = node.parent
        right = node.right
        # create new split node
        tree = Node('{feature} {operator} {value}'.format(feature = feature,operator = operator,value = value),parent = parent, right = right, checked = False)
        # Split dataset on decision
        leftDataFrame, rightDataFrame = split_data(feature, operator, value, dataFrame)

        if(not leftDataFrame.empty):
            create_leaf(leftDataFrame,parent=tree,right=False)
        if(not rightDataFrame.empty):
            create_leaf(rightDataFrame,parent=tree,right=True)
        # create child nodes of decision
        #tree.children = (create_leaf(leftDataFrame,parent=tree,right=False),create_leaf(rightDataFrame,parent=tree,right=True))
        node.parent = None
        # Place left nodes back on left
        place_left_node(tree)


        # change node to decision
        return tree
    else:
        return node

    
    
# Delete the split at the given node
# In place function
def delete_split(node, dataFrame):
    # get the parent node
    parent = node.parent
    # get the position of the node relative to the parent
    right = node.right
    
    # Run data through tree to this point
    dataFrame = data_values(node,dataFrame)
    
    newNode = create_leaf(dataFrame, parent, right)
    # Remove original node from tree
    node.parent = None
    place_left_node(newNode)
    
    return newNode

# Create a new split at the node
def update_split(node, dataFrame):
    # find optimal split
    # replace previous split with new optimal split
    return optimal_split(node,dataFrame)

def create_split(node, dataFrame):
    if node.is_leaf:
        if node.accuracy < 1:
            optimal_split(node,dataFrame)
    else:
        optimal_split(node,dataFrame)
            

def get_random_node(tree):
    availableNodes = [node for node in PreOrderIter(tree) if not node.checked]
    if availableNodes:
        return random.choice(availableNodes)
    else:
        return None

# gets root of node
# if no root, returns node
def get_root(node):
    if(node.root):
        return node.root
    else:
        return node

def random_node_modification(tree, dataFrame, n_min = 1, D_max = 5, alpha = 0.5, n_treatment = 10, mu = 0.5):
    improvement = 1
    while get_random_node(tree):
        totalLoss = loss_function(tree, dataFrame, alpha)
        node = get_random_node(tree)
        loss = loss_function(node, dataFrame, alpha)
        #print("New loop, current loss: ",totalLoss)
        #print("Chosen Node:", node, " Loss at node: ", )
        
        if node.children:
            testNode = copy.deepcopy(node)
            newNode = delete_split(testNode, dataFrame)
            newLoss = loss_function(newNode,dataFrame, alpha)
            if newLoss < loss:
                tree = copy.deepcopy(get_root(newNode))
                # reset improvement counter
                improvement = 0
            testNode = copy.deepcopy(node)
            newNode = update_split(testNode, dataFrame)
            newLoss = loss_function(newNode,dataFrame, alpha)
            if newLoss < loss:
                tree = copy.deepcopy(get_root(newNode))
                #reset improvement counter
                improvement = 0
        else:
            testNode = copy.deepcopy(node)
            testTree = get_root(testNode)
            create_split(testNode, dataFrame)
            newLoss = loss_function(testTree, dataFrame, alpha)
            if newLoss < loss:
                tree = copy.deepcopy(testTree)
                # reset improvement counter
                improvement = 0
        if improvement > 0:
            node.checked = True
        improvement += 1
    return tree
```

```python
def error_count(tree, dataFrame):
    error = 0
    # misclassification error
    if not tree.is_leaf:
        # split data by decision
        feature, operator, value = tree.name.split()
        leftData,rightData  = split_data(feature, operator, value, dataFrame)
        
        # check if node is on left
        if((not leftData.empty) and not tree.children[0].right):
            error += error_count(tree.children[0],leftData)
        if((not rightData.empty) and len(tree.children)>1):
            error += error_count(tree.children[1],rightData)
        elif((not rightData.empty) and tree.children[0].right):
            error += error_count(tree.children[0],rightData)
    else:

        '''
        minimising with respect to tau(.):
        u * sum of (outcome of point if leaf has the same prescription + sum of rest of the means of the other prescriptions)
        + (1-u) * sum of (outcome of point - mean of outcome of same treatment of that leaf)**2

        '''
        
        error += len(dataFrame.query("Class != @tree.Class").index)
    return error
test_data['Class'].value_counts().iloc[0]/len(test_data['Class'].index)
```
